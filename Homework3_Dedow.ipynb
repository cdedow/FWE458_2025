```python
import pandas as pd #importing the pandas library

fname = filedir + "MaunaLoaCO2_monthly_sitename.csv" #preparing to load into data frame

df = pd.read_csv(fname) #loaded the file into a pandas data frame
df.head() #displayed the first 5 rows
```

```python
#Filter the data: exclude years 1958 and 2022
df = df[(df["Year"] != 1958) & (df["Year"] != 2022)] #made a new data frame that kept all of the rows with all the years excpet 1958 and 2022 using the != "does not equal".
print(df.head()) #displayed the first 5 rows of this new data frame, showing that all of the years as 1958 was taken out
print(df.tail()) #displated the last 5 rows of this new data frame, showing that all of the years as 2022 was taken out
```

```python
#Filter the data: remove specific rows with invalid data
df = df[(df["numofdays"] > 0) & (df["stdofdays"] > 0)] #created a new data frame that kept rows that only had positive values in the number of days and standard deviation of days
df.head() #displated the new data frame showing that they are only positive
```

```python
#Compute annual statistcs: annual average co2 concentration
annual_average = df.groupby("Year")["Monthly_average"].mean() #created a new variable that stores a series that groups the previous data frame into each year and then finds the mean of the monthly average column
annual_average #displaying the variable
```

```python
#Compute annual statistcs: annual average deseasonalized co2 concentration
deseasonalized_annual_average = df.groupby("Year")["deseasonalized"].mean() #created a new variable that stores a series that groups the data frame into each year and then finds the mean of the deseasonalized column
deseasonalized_annual_average #displaying the variable
```

```python
#Compute annual statistcs: annual average deseasonalized co2 concentration
std_of_months = df.groupby("Year")["Monthly_average"].std() #created a new variable that stores a series that groups the data frame into each year and then finds the standard deviation of the monthly average column
std_of_months #displaying the variable
```

```python
#define custom function
def categorize_co2(value): #define and give the function a name and input space
  if value < 350: #if statement checks if input is less than 350
    return "Low" #it will return low
  elif value < 400: #else if statement checks if input is less than 400
    return "Medium" #it will return medium
  else: #anything else (greater than 400)
    return "High" #it will return high
```

```python
#apply the custom function to make a new column
df["CO2_category"] = df["Monthly_average"].apply(categorize_co2) #edits the data frame, adding a new column that applies the new function to the monthly average column and puts out the return in that column
df.head() #displays part of the data frame
```
